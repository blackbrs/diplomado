```python
import pandas as pd
import datetime
#Trasacciones del genero Masculino
t_m = pd.read_csv('TransaccionesM.csv')
t_m = t_m.drop(t_m.columns[0], axis=1)
t_m
```

```python
#transacciones del genero femenino
t_f= pd.read_csv('TransaccionesF.csv')
t_f = t_f.drop(t_f.columns[0], axis=1)
t_f
```

```python
#cambiando el formato de fecha a los dataframe
#Dataframe Masculino
t_m['Birthdate'] = pd.to_datetime(t_m['Birthdate'],format='%Y-%m-%d')
t_m['Birthdate'] = t_m['Birthdate'].dt.strftime('%d/%m/%Y')
t_m['Date'] = pd.to_datetime(t_m['Date'],format='%Y-%m-%d')
t_m['Date'] = t_m['Date'].dt.strftime('%d/%m/%Y')
t_m
```

```python
#Dataframe Femenino
t_f['Birthdate'] = pd.to_datetime(t_f['Birthdate'],format='%Y-%m-%d')
t_f['Birthdate'] = t_f['Birthdate'].dt.strftime('%d/%m/%Y')
t_f['Date'] = pd.to_datetime(t_f['Date'],format='%Y-%m-%d')
t_f['Date'] = t_f['Date'].dt.strftime('%d/%m/%Y')
t_f
```

```python
#Hacemos la concatenacion de los dos Dataframes
consolidado_transacciones = pd.concat([t_m,t_f], axis=0)
consolidado_transacciones
```

```python
#Generar un tercer dataframe con una llave comun quepermita unificar o  mezclar
df_index = consolidado_transacciones.set_index('Category')
df_index
```
